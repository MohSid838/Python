```python
#using math function
import math
class Solution():
    def sqt(self,x:int)->int:
        if x%2==0:
            return int(pow(x,0.5))
        else: None
print(Solution().sqt(8))
```

```python
#not using math function:
class Solution():
    def mysqrt(self,x:int)->int:
        start=1
        end=x
        result=0
        while start<=end:
            mid=start+(end-start)//2
            if mid*mid==x:
                result-mid
                return result
            elif mid*mid>x:
                end=mid-1
            else:
                result=mid
                start=mid+1
        return result
print(Solution().mysqrt(8))
```

Binary Search

Problem Statement :

Given an array of integers nums which is sorted in ascending order, and an integer target, write a function to search target in nums. If target exists, then return its index. Otherwise, return -1.

Examples:
 

Example 1:

Input: nums = [-1,0,3,5,9,12], target = 9
Output: 4
Explanation: 9 exists in nums and its index is 4.
 

Example 2:

Input: nums = [-1,0,3,5,9,12], target = 2
Output: -1
Explanation: 2 does not exist in nums so return -1.

```python
from typing import List
class Solution:
    def binsearch(self,x:int,nums:List[int])->int:
        start=0
        end=len(nums)-1
        while start <=end:
            mid=(start+end)//2
            if nums[mid]==x:
                return mid
            elif nums[mid] < x:
                start=mid+1
            else:
                end=mid-1
        return -1   # not found
print(Solution().binsearch(5, [1,2,3,4,5,6,7]))
```

```python
from typing import List
class Solution:
    def intersect(self,nums1:List[int],nums2:List[int])->List[int]:
        nums1.sort()
        nums2.sort()
        i,j=0,0
        result=[]
        while i < len(nums1) and j < len(nums2):
            if nums1[i] < nums2[j]:
                i += 1
            elif nums1[i] > nums2[j]:
                j += 1
            else:
                result.append(nums1[i])
                i += 1
                j += 1
        return result
print(Solution().intersect([1,2,2,1], [2,2]))
```

```python
class Solution:
    def negativenumbers(self,grid):
        negative=0
        for i in range(len(grid)):
            start=0
            end=len(grid[0])-1
            while start <=end:
                mid=start+(end-start)//2
                if grid[i][mid]<0:
                    end=mid-1
                else:
                    start=mid+1
            #print(end)
            negative+=abs(len(grid[0])-end-1)
        return negative
print(Solution().negativenumbers([[4,3,2,-1],[3,2,1,-1],[1,1,-1,-2],[-1,-1,-2,-3]]))
print(Solution().negativenumbers([[3,2],[1,0]]))
```

```python
class Solution:
    def sortorderlist(self,nums):
        c=0
        for i in range(len(nums)):
            if nums[i]>nums[(i+1)%len(nums)]:
                c+=1
            if c>1:
                return False
        return True
print(Solution().sortorderlist([3,4,5,1,2]))    
```

```python
class Solution:
    def verifylist(self,s:str,goal:str)->bool:
        if len(s)!=len(goal):
            return False
        goal+=goal
        return s in goal
print(Solution().verifylist("abcde","cdeab"))
print(Solution().verifylist("abcde","abced"))
```

```python
from typing import List
class Solution:
    def findPeakelement(self,nums:List[int])->int:
        low=0
        high=len(nums)-1
        while low<high:
            mid=low+(high-low)//2
            if nums[mid]>nums[mid+1]:
                high=mid
            else:
                low=mid+1
        return low
print(Solution().findPeakelement([1,2,3,1]))
print(Solution().findPeakelement([1,2,1,3,5,6,4]))
```

```python
class Solution:
    def singleNonDuplicate(self, nums: List[int]) -> int:
        low = 0
        high = len(nums) - 1
        while low < high:
            mid = (low + high) // 2
            if mid % 2 == 1:
                mid -= 1
            if nums[mid] == nums[mid + 1]:
                low = mid + 2
            else:
                high = mid
        return nums[low]
print(Solution().singleNonDuplicate([1,1,2,3,3,4,4,8,8]))
print(Solution().singleNonDuplicate([3,3,7,7,10,11,11]))
```
